```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#Question 2a
x = np.random.uniform(0,1,(30)).reshape(30,1)
epsilon = np.random.normal(0, 0.07,30).reshape(30,1)
gx = ((np.sin(2*np.pi*x))**2) + epsilon 
```

```python
xrange = np.arange(0,1,0.01)
sin = np.sin(2*np.pi*xrange)**2
```

```python
plt.plot(x,gx, 'bo', xrange, sin)
plt.axis([0,1,0,1.5])
```

```python
#function for fitting model of polynomial basis k to dataset, returns array of parameters
def fit(k, x, y):
    A = []
    
    for i in range(k):
        A.append(x**i)
        
    A = np.concatenate(A, axis = 1)
    
    #smallest indices first, largest indices last
    params = np.linalg.inv(A.T@A)@(A.T@y)
    return params
        
```

```python
#K =2 
P2 = fit(2,x, gx)

a2 = P2[1][0]
b2 = P2[0][0]

yhat2 = a2*(xrange) + b2
```

```python
#K=5
P5 = fit(5,x, gx)

a5 = P5[4][0]
b5 = P5[3][0]
c5 = P5[2][0]
d5 = P5[1][0]
e5 = P5[0][0]

yhat5 = a5*(xrange**4) + b5*(xrange**3) + c5*(xrange**2) + d5*(xrange) + e5 
```

```python
#K=10
P10 = fit(10,x, gx)


a10 = P10[9][0]
b10 = P10[8][0]
c10 = P10[7][0]
d10 = P10[6][0]
e10 = P10[5][0]
f10 = P10[4][0]
g10 = P10[3][0]
h10 = P10[2][0]
i10 = P10[1][0]
j10 = P10[0][0]

yhat10 = a10*(xrange**9) + b10*(xrange**8) + c10*(xrange**7) + d10*(xrange**6) + e10*(xrange**5) + f10*(xrange**4) + g10*(xrange**3) + h10*(xrange**2) + i10*xrange + j10
```

```python
#K=14
P14 = fit(14,x, gx)


a14 = P14[13][0]
b14 = P14[12][0]
c14 = P14[11][0]
d14 = P14[10][0]
e14 = P14[9][0]
f14 = P14[8][0]
g14 = P14[7][0]
h14 = P14[6][0]
i14 = P14[5][0]
j14 = P14[4][0]
k14 = P14[3][0]
l14 = P14[2][0]
m14 = P14[1][0]
n14 = P14[0][0]


yhat14 = a14*(xrange**13) + b14*(xrange**12) + c14*(xrange**11) + d14*(xrange**10) + e14*(xrange**9) + f14*(xrange**8) + g14*(xrange**7) + h14*(xrange**6) + i14*(xrange**5) + j14*(xrange**4) + k14*(xrange**3) + l14*(xrange**2) + m14*xrange + n14
```

```python
#K=18
P18 = fit(18,x, gx)

a18 = P18[17][0]
b18 = P18[16][0]
c18 = P18[15][0]
d18 = P18[14][0]
e18 = P18[13][0]
f18 = P18[12][0]
g18 = P18[11][0]
h18 = P18[10][0]
i18 = P18[9][0]
j18 = P18[8][0]
k18 = P18[7][0]
l18 = P18[6][0]
m18 = P18[5][0]
n18 = P18[4][0]
o18 = P18[3][0]
p18 = P18[2][0]
q18 = P18[1][0]
r18 = P18[0][0]

yhat18 = a18*(xrange**17) + b18*(xrange**16) + c18*(xrange**15) + d18*(xrange**14) + e18*(xrange**13) + f18*(xrange**12) + g18*(xrange**11) + h18*(xrange**10) + i18*(xrange**9) + j18*(xrange**8) + k18*(xrange**7) + l18*(xrange**6) + m18*(xrange**5) + n18*(xrange**4) + o18*(xrange**3) + p18*(xrange**2) + q18*xrange + r18
```

```python
plt.plot(x,gx, 'ro', xrange,yhat2, xrange,yhat5, xrange,yhat10, xrange,yhat14, xrange,yhat18)
plt.axis([0,1,0,1.5])
```

```python
#Question 2b

yhat2 = a2*(x) + b2
yhat5 = a5*(x**4) + b5*(x**3) + c5*(x**2) + d5*(x) + e5 
yhat10 = a10*(x**9) + b10*(x**8) + c10*(x**7) + d10*(x**6) + e10*(x**5) + f10*(x**4) + g10*(x**3) + h10*(x**2) + i10*x + j10
yhat14 = a14*(x**13) + b14*(x**12) + c14*(x**11) + d14*(x**10) + e14*(x**9) + f14*(x**8) + g14*(x**7) + h14*(x**6) + i14*(x**5) + j14*(x**4) + k14*(x**3) + l14*(x**2) + m14*x + n14
yhat18 = a18*(x**17) + b18*(x**16) + c18*(x**15) + d18*(x**14) + e18*(x**13) + f18*(x**12) + g18*(x**11) + h18*(x**10) + i18*(x**9) + j18*(x**8) + k18*(x**7) + l18*(x**6) + m18*(x**5) + n18*(x**4) + o18*(x**3) + p18*(x**2) + q18*x + r18
```

```python
def MSEnp(y,yhat,n):
    sqrdiff = (y - yhat)**2
    SSE = np.sum(sqrdiff)
    meanSE = SSE/n
    return meanSE
```

```python
MSE = [np.log(MSEnp(gx,yhat2,30)), np.log(MSEnp(gx,yhat5,30)), np.log(MSEnp(gx,yhat10,30)), np.log(MSEnp(gx,yhat14,30)),np.log(MSEnp(gx,yhat18,30))]
print(MSE)
K = [2,5,10,14,18] 

plt.plot(K, MSE, 'bo')
plt.xlabel('polynomial basis dimension K')
plt.ylabel('MSE train (ln)')
```
